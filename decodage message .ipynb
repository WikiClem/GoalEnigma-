```python
from math import ceil
import re

msg = open("S1/message1.txt", "r", encoding="utf-8").read()
n = len(msg)
W = set("le la les un une des de du ce cette je tu il elle nous vous on qui que et ou mais est dans sur avec pour par ne pas plus si tout comme très bien être avoir sans sont fait faire dit dire".split())

def dec(k):
    r, f = ceil(n/k), n%k or k
    g, p = [], 0
    for c in range(k):
        l = r if c < f else r-1; g.append(msg[p:p+l]); p += l
    return "".join(g[c][i] for i in range(r) for c in range(k) if i < len(g[c]))

sc = lambda  t: sum(m in W for m in re.findall(r'[a-zàâäéèêëïîôùûüç]+', t.lower()))
best = max(range(2, len(msg)), key=lambda k: sc(dec(k)))
t = dec(best)
print(f"Colonnes: {best} | Score: {sc(t)}\n\n{t}")
```

```python
import re

# Lecture du message chiffré
msg2 = open("S1/message2.txt", "r", encoding="utf-8").read()

# Liste de mots français courants pour évaluer si le décodage est correct
mots_fr = {
    "le", "la", "les", "un", "une", "des", "de", "du", "ce", "cette", "ces",
    "mon", "ton", "son", "notre", "votre", "leur",
    "je", "tu", "il", "elle", "nous", "vous", "ils", "elles", "on",
    "qui", "que", "quoi", "dont", "où", "et", "ou", "mais", "donc",
    "est", "sont", "être", "avoir", "fait", "faire",
    "dans", "sur", "avec", "sans", "pour", "par",
    "bien", "très", "plus", "moins", "ne", "pas", "tout", "comme", "si",
    "bonjour", "message", "code", "texte"
}

def decoder_cesar(texte, decalage):
    """
    Décode un texte chiffré avec un décalage César.
    Chaque caractère est décalé de 'decalage' positions dans la table Unicode.
    """
    resultat = ""
    for caractere in texte:
        nouveau_code = ord(caractere) - decalage
        # Vérifier que le code Unicode est valide
        if 0 <= nouveau_code < 0x110000:
            resultat += chr(nouveau_code)
        else:
            resultat += caractere  # Garder le caractère original si invalide
    return resultat

def calculer_score(texte):
    """
    Calcule un score basé sur le nombre de mots français reconnus dans le texte.
    Plus le score est élevé, plus le texte ressemble à du français.
    """
    # Extraire tous les mots (lettres minuscules, avec accents)
    mots_trouves = re.findall(r'[a-zàâäéèêëïîôùûüç]+', texte.lower())
    # Compter combien de mots sont dans notre dictionnaire
    score = sum(1 for mot in mots_trouves if mot in mots_fr)
    return score

# Tester tous les décalages possibles de -200 à +200
meilleur_decalage = 0
meilleur_score = 0
meilleur_texte = ""

for decalage in range(-200, 201):
    texte_decode = decoder_cesar(msg2, decalage)
    score = calculer_score(texte_decode)
    
    if score > meilleur_score:
        meilleur_score = score
        meilleur_decalage = decalage
        meilleur_texte = texte_decode

# Afficher le résultat
print(f"Décalage trouvé: {meilleur_decalage}")
print(f"Score: {meilleur_score} mots français reconnus")
print(f"\nMessage décodé:\n{meilleur_texte}")
```
