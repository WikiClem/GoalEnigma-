```python
from math import ceil
import re

msg = open("S1/message1.txt", "r", encoding="utf-8").read()
n = len(msg)
W = set("le la les un une des de du ce cette je tu il elle nous vous on qui que et ou mais est dans sur avec pour par ne pas plus si tout comme très bien être avoir sans sont fait faire dit dire".split())

def dec(k):
    r, f = ceil(n/k), n%k or k
    g, p = [], 0
    for c in range(k):
        l = r if c < f else r-1; g.append(msg[p:p+l]); p += l
    return "".join(g[c][i] for i in range(r) for c in range(k) if i < len(g[c]))

sc = lambda  t: sum(m in W for m in re.findall(r'[a-zàâäéèêëïîôùûüç]+', t.lower()))
best = max(range(2, len(msg)), key=lambda k: sc(dec(k)))
t = dec(best)
print(f"Colonnes: {best} | Score: {sc(t)}\n\n{t}")
```

```python
import re

# Lecture du message chiffré
msg2 = open("S1/message2.txt", "r", encoding="utf-8").read()


def decoder_cesar(texte, decalage):
    """
    Décode un texte chiffré avec un décalage César.
    Chaque caractère est décalé de 'decalage' positions dans la table Unicode.
    """
    resultat = ""
    for caractere in texte:
        nouveau_code = ord(caractere) - decalage
        # Vérifier que le code Unicode est valide
        if 0 <= nouveau_code < 0x110000:
            resultat += chr(nouveau_code)
        else:
            resultat += caractere  # Garder le caractère original si invalide
    return resultat

def calculer_score(texte):
    """
    Calcule un score basé sur le nombre de mots français reconnus dans le texte.
    Plus le score est élevé, plus le texte ressemble à du français.
    """
    # Extraire tous les mots (lettres minuscules, avec accents)
    mots_trouves = re.findall(r'[a-zàâäéèêëïîôùûüç]+', texte.lower())
    # Compter combien de mots sont dans notre dictionnaire
    score = sum(1 for mot in mots_trouves if mot in mots_fr)
    return score

# Tester tous les décalages possibles de -200 à +200
meilleur_decalage = 0
meilleur_score = 0
meilleur_texte = ""

for decalage in range(-200, 201):
    texte_decode = decoder_cesar(msg2, decalage)
    score = calculer_score(texte_decode)
    
    if score > meilleur_score:
        meilleur_score = score
        meilleur_decalage = decalage
        meilleur_texte = texte_decode

# Afficher le résultat
print(f"Décalage trouvé: {meilleur_decalage}")
print(f"Score: {meilleur_score} mots français reconnus")
print(f"\nMessage décodé:\n{meilleur_texte}")
```

```python
import re

# Lecture du message chiffré
msg = open("message 4.txt", "r", encoding="utf-8").read()

# Dictionnaire de mots français pour évaluer le décodage
mots_fr = {"le", "la", "les", "un", "une", "des", "de", "du", "ce", "cette", "ces",
           "je", "tu", "il", "elle", "nous", "vous", "on", "qui", "que", "et", "ou",
           "mais", "est", "dans", "sur", "avec", "pour", "par", "ne", "pas", "plus",
           "si", "tout", "comme", "très", "bien", "être", "avoir", "sans", "sont",
           "fait", "faire", "dit", "dire", "message", "code", "texte", "alphabet"}

# Séparer les caractères pairs (0,2,4...) et impairs (1,3,5...)
chars_pairs = msg[0::2]      # Positions 0, 2, 4, 6...
chars_impairs = msg[1::2]    # Positions 1, 3, 5, 7...

def decoder_cesar(texte, decalage):
    """Décode un texte avec un décalage César"""
    resultat = ""
    for c in texte:
        nouveau_code = ord(c) - decalage
        if 0 <= nouveau_code < 0x110000:
            resultat += chr(nouveau_code)
        else:
            resultat += c
    return resultat

def recombiner(a, b):
    """Recombine deux chaînes en alternant leurs caractères"""
    resultat = ""
    for i in range(min(len(a), len(b))):
        resultat += a[i] + b[i]
    # Ajouter le dernier caractère si a est plus long
    if len(a) > len(b):
        resultat += a[-1]
    return resultat

def calculer_score(texte):
    """Compte le nombre de mots français reconnus"""
    mots = re.findall(r'[a-zàâäéèêëïîôùûüç]+', texte.lower())
    return sum(1 for mot in mots if mot in mots_fr)

# Recherche des meilleures clés
meilleur_score = 0
meilleure_cle = (0, 0)

for cle_pair in range(-100, 101):
    for cle_impair in range(-100, 101):
        # Décoder chaque partie avec sa clé
        decode_pair = decoder_cesar(chars_pairs, cle_pair)
        decode_impair = decoder_cesar(chars_impairs, cle_impair)
        
        # Recombiner et calculer le score
        texte_decode = recombiner(decode_pair, decode_impair)
        score = calculer_score(texte_decode)
        
        # Garder le meilleur résultat
        if score > meilleur_score:
            meilleur_score = score
            meilleure_cle = (cle_pair, cle_impair)

# Afficher le résultat final
texte_final = recombiner(
    decoder_cesar(chars_pairs, meilleure_cle[0]),
    decoder_cesar(chars_impairs, meilleure_cle[1])
)

print(f"Clé positions paires: {meilleure_cle[0]}")
print(f"Clé positions impaires: {meilleure_cle[1]}")
print(f"Score: {meilleur_score} mots reconnus")
print(f"\nMessage décodé:\n{texte_final}")
```
